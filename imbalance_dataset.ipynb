## Handling Imbalance Dataset

What is an Imbalanced Dataset?

When one class dominates the other.

📌 Example:
95% → Class 0
5% → Class 1

👉 Model may just predict Class 0 always and still get 95% accuracy (misleading ❌)

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(24)

n_samples = 1000
class_0_ratio = 0.9

n_class_0 = int(n_samples*class_0_ratio)
n_class_1 = n_samples - n_class_0
```

```python
n_class_0,n_class_1
```

```python
# Creating Dataframe with imbalanced dataset

class_0 = pd.DataFrame({
    "feature_1": np.random.normal(loc=0,scale=1,size=n_class_0),
     "feature_2": np.random.normal(loc=0,scale=1,size=n_class_0),
     "target": [0]*n_class_0
})

class_1 = pd.DataFrame({
    "feature_1": np.random.normal(loc=2,scale=1,size=n_class_1),
     "feature_2": np.random.normal(loc=2,scale=1,size=n_class_1),
     "target": [1]*n_class_1
})
```

```python
df = pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df["target"].value_counts()
```

#### 🔥 Techniques to Handle Imbalanced Data

##### Oversampling (Increase minority class)

```python
df_minority = df[df["target"]==1]
df_majority = df[df["target"]==0]
```

```python
! pip install scikit-learn
```

```python
from sklearn.utils import resample

df_minority_resampled = resample(
    df_minority,replace=True,
    n_samples=len(df_majority),
    random_state=42
)
```

```python
df_minority_resampled.shape
```

```python
df_upsampled = pd.concat([df_majority,df_minority_resampled])
```
